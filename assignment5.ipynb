Questions:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
s=pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv')
s.head()
```

```python
s.shape
```

```python
#1.Delete unnamed columns
s.drop(columns='Unnamed: 0',axis=1)
```

```python
#2. Show the distribution of male and female
malefemale_distribution=s.pivot_table(index='Gender',values='Count',aggfunc=np.sum)
malefemale_distribution
```

```python
print("The Distribution of male and female ")


labels = 'Female', 'Male'
explode = (0.1, 0)  # explode the 1st slice


plt.pie(malefemale_distribution, explode=explode, labels=labels, autopct='%1.02f%%', startangle=90,shadow=True)


plt.axis('equal') # making the pie chart flat
plt.title("Male_Female Distribution")
```

```python
#3. Show the top 5 most preferred names
print(" the top 5 most preferred names in the given data is:")

top5_names = s.groupby(['Name'])['Count'].sum().nlargest(5)
top5_names
```

```python
#4. What is the median name occurence in the dataset
print('The median name occurence in the dataset is:')
s[s.Id == s.median().Id]['Name']
```

```python
#5. Distribution of male and female born count by states
print('The distribution of male and female born count by states is:')

genders =s.pivot_table(index='State',columns='Gender', values='Count',aggfunc=np.sum)
genders.head()
```

# Task 2

We have the min and max temperatures in a city In India for each months of the year. We
would like to find a function to describe this and show it graphically, the dataset given below.
Task:
1. fitting it to the periodic function
2. plot the fit
Data
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
import numpy as np
import matplotlib.pyplot as plt

temp_max = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25]) # Fitting in periodic function
temp_min = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18]) # Fitting in periodic function

months = np.arange(12)

plt.plot(months, temp_max, 'ro')
plt.plot(months, temp_min, 'bo')

plt.xlabel('Month')
plt.ylabel('Min and Max Temperature')
```

```python
from scipy import optimize

def yearly_temps(times, avg, ampl, time_offset):
    return (avg + ampl*np.cos((times+time_offset) * 2 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temps, months, temp_max, [20,10,0])  # Plotting the fit
res_min, cov_min = optimize.curve_fit(yearly_temps, months, temp_min, [-40,20,0]) # Plotting the fit
```

```python
days = np.linspace(0, 12 , num=365)

plt.figure()

plt.plot(months, temp_max, 'ro')
plt.plot(days, yearly_temps(days, *res_max), 'r-')

plt.plot(months, temp_min, 'bo')
plt.plot(days, yearly_temps(days, *res_min), 'b-')

plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')

plt.show()
```
